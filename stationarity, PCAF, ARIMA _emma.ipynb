```python
#source: https://www.analyticsvidhya.com/blog/2018/09/non-stationary-time-series-python/

def adf_test(timeseries):
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
       dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

# Learning more about the time series (ADF/ACF/PACF)

Now that we have the data parsed, we're going to work towards modeling this data using a common method in statistics: ARIMA. But to do that, we need to learn about the data a little bit first.

First, we need to learn if the data is stationary. If it's not, we need to incorporate some type of integration component in our model. Put another way, if the data has a mean that changes across time, we're going to need to do something about that later in our ARIMA.

```python
adf_test(googletrends['chipotle: (United States)'])
```

The p here is > .05. The null hypothesis for both tests is that the data are non-stationary. We want to REJECT the null hypothesis for this test, so we want a p-value of less that 0.05 (or smaller). Our p > .05, so the mean appears to change across time. We're going to want to build a model that removes this stationarity from our data before we model seasonality and autoregression.

Another way to remove stationarity from a time series is to difference it. Let's take a look at what differencing is.

```python
googletrends['trends_diff_1'] = googletrends['chipotle: (United States)'].diff(1)
```

```python
googletrends
```

A difference is just the change from one time measurement to another. The change from 88 to 86 == -2. The change from 86 to 86 == 0, and so on. First order differencing is just taking the last measurement, but you can difference multiple time periods at once. Typically, 2nd, 3rd and 4th order differencing are only done if first order differencing doesn't make the data stationary.

```python
pd.plotting.register_matplotlib_converters()
sns.set(rc={'figure.figsize':(14, 4)})
googletrends['trends_diff_1'].plot(linewidth=0.75)
```

Do you see a moving average, or general trend, here now? Me neither.

```python
adf_test(googletrends['trends_diff_1'][1:])
# now pvalue is significant so we can reject the null and say that the data has stationarity 
#The null hypothesis for both tests is that the data are non-stationary#
```

Hey, why am I doing [1:]? 

Our p value is really low, meaning we're as stationary as all of those $GME 
stockholders.

If we're going to build an ARIMA model, we need to know a bit about how the data tends to fluctuate consistently across time. We're going to manually model these behaviors using time series components. These are called the lag values for the Autoregression (AR) and Moving Average (MA) parameters, or p and q parameters for an ARIMA model, respectively.

The autocorrelation plot checks the degree to which past datapoints inform future ones. A lag of 1 days says that the value from yesterday informs the value of this week.

```python
plot_acf(googletrends['trends_diff_1'][1:], lags=53);
```

If the lag effect goes beyond the blue shaded region, it's worth considering for a model. For this data, I'm seeing AR(1), AR(2), AR(4) and possibly AR(6) and AR(30).

Now, let's think about partial autocorrelation.

The moving average process is an autoregression model of the time series of residual errors from what we've already modeled with AR.

Another way to think about the moving average model is that it corrects future forecasts based on errors made on recent forecasts.

More recent trends that are not subject to the same types of calendar effects (e.g., day of week, month, so on).

```python
plot_pacf(googletrends['trends_diff_1'][1:], lags=31)
# plotting the residuals to see if the errors have a correlation with itself that we are left with (unexplained variance)
```

Possibly MA(1), MA(2), MA(3). That is, the residual error could be correlated up to 3 days.

# SARIMA Modeling


```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Taking the parameters we learned from ADF, ACF and PACF, let's try a model. Note that SARIMA allows data with general trends, so we don't need to make the data stationary, but instead feed a trend parameter. Remember, we're trying to beat an AIC of 2524.816

```python
model = sm.tsa.statespace.SARIMAX(googletrends['chipotle: (United States)'], trend='t', order=(3,0,3))
model_fit = model.fit(disp=False)
print(model_fit.summary())
```

This is a pretty good model! Is it better than your ARIMA model? No, it is not.

Let's add a seasonal component now. There's very little documentation on how seasonality works in an SARIMA model (via the seasonal_order), but know that in general, the last parameter corresponds to the type of seasonality we'd expect. We saw some longer autoregressive significance at lag 28, so lets try that. The first three parameters refer to AR I and MA components again. Think of the AR and MA components as the number of past seasonality points that you would expect to see a seasonal effect. 

So let's start here with the idea of a 7 day (weekly) seasonality. We'll feed the model (4,0,4) because we expect to see this effect up to four weeks out.

```python
model = sm.tsa.statespace.SARIMAX(googletrends['chipotle: (United States)'], trend='t', order=(2,1,3), seasonal_order=(1,0,1,4))
model_fit = model.fit(disp=False)
print(model_fit.summary())
```

This is a completely overfit and non-significant model. Notice how the AR and MA compoenets now become insignificant. Let's try removing the the regular AR and MA components to see if seasonality predicts the data better than a traditional ARIMA approach.

```python
model = sm.tsa.statespace.SARIMAX(googletrends['chipotle: (United States)'], trend='t', seasonal_order=(1,0,1,7))
model_fit = model.fit(disp=False)
print(model_fit.summary())
```

Good! They are significant, at least in a model with no AR and MA components. The AIC doesn't beat our best ARIMA model, though.

Looking better! AR(1) isn't significant, but AR(2) and AR(3) are, so let's leave that parameter where it is. Unfortunately it's not easily to selectively pick AR(2) and AR(3) in a model. The seasonal MA component is not significant, so let's remove that.

```python
model = sm.tsa.statespace.SARIMAX(googletrends['chipotle: (United States)'], trend='t', order=(3,0,3), seasonal_order=(1,0,0,7))
model_fit = model.fit(disp=False)
print(model_fit.summary())
```

Beautiful. The AR lags became more significant, and so did the seasonal lags. However, this AIC doesn't beat our ARIMA model, so seasonality doesn't explain unique variance in this model, as far as I can tell.

```python
residuals = pd.DataFrame(model_fit.resid)
sns.distplot(residuals, hist=True, kde=True)
```

```python
# creating regressors and target variable for training and validation
def makeXy(ts, nb_timesteps):
    """
    Input: 
           ts: original time series
           nb_timesteps: number of time steps in the regressors
    Output: 
           X: 2-D array of regressors
           y: 1-D array of target 
    """
    X = []
    y = []
    for i in range(nb_timesteps, ts.shape[0]):
        X.append(list(ts.loc[i-nb_timesteps:i-1]))
        y.append(ts.loc[i])
    X, y = np.array(X), np.array(y)
    return X, y
```

```python
#Define input layer which has shape (None, 7) and of type float32. None indicates the number of instances
input_layer = Input(shape=(7,1), dtype='float32')
```

```python
#LSTM layer is defined for seven timesteps
gru_layer = GRU(64, input_shape=(7,1), return_sequences=False)(input_layer)
```

```python
dropout_layer = Dropout(0.2)(gru_layer)
```

```python
#Finally the output layer gives prediction for the next day's air pressure.
output_layer = Dense(1, activation='linear')(dropout_layer)
```

```python
ts_model = Model(inputs=input_layer, outputs=output_layer)
ts_model.compile(loss='mse', optimizer='adam')
ts_model.summary()
```

```python

"""
The model is trained by calling the fit function on the model object and passing the X_train and y_train. The training 
is done for a predefined number of epochs. Additionally, batch_size defines the number of samples of train set to be
used for a instance of back propagation.The validation dataset is also passed to evaluate the model after every epoch
completes. A ModelCheckpoint object tracks the loss function on the validation set and saves the model for the epoch,
at which the loss function has been minimum.
"""
save_weights_at = os.path.join('keras_models', 'chipotle_GRU_weights.{epoch:02d}-{val_loss:.4f}.hdf5')
save_best = ModelCheckpoint(save_weights_at, monitor='val_loss', verbose=0,
                            save_best_only=True, save_weights_only=False, mode='min',
                            period=1)
ts_model.fit(x=X_train, y=y_train, batch_size=16, epochs=20,
             verbose=1, callbacks=[save_best], validation_data=(X_val, y_val),
             shuffle=True)
```
